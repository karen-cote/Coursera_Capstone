## Introduction / Business Problem

Food deserts are a well researched problem in public health.  The availability of grocery stores with good food choices within a mile (urban) or 10 miles (rural) of a person's home have a significant impact on health and the ability to eat nutritious meals.   

This project takes the food desert idea and looks at a smaller population - high school students.  Schools often allow students to leave campus for their lunch period, particularly seniors, who may not have full schedules.  Students also look for food after school or after practice.  Their choices nearby influence their diet - and if the only choices are high-fat, high-sugar fast foods, that diet may be subject to less than ideal inputs.  In a high poverty area, quick choices may also replace nutrituous home cooked meals if the area is also a food desert.  So all the pieces tie together when looking at diet overall.

By mapping high school locations, along with food options within a mile radius, it will be possible to analyze the nearby choices for each high school.  By clustering the similar option areas together, we can highlight areas where there are more and less choices, and more and less nutritious ones as well.  

The outcome of this analysis could be used in two ways - to help public health and primary care practitioners (pediatricians, primarily) understand what their patients may be eating, and how those available options might translate into their patient population (allergies, asthma, obesity). The analysis could help inform this group how to target interventions for their patients - suggestions for where and what to eat, as an example, to result in better health, and better school and athletic performance.

The other use of this data is business related - finding the types of food that is already available can help define possible new options to fill an open niche.  TOgether, the two uses create a powerful partnership that could bring options into neighborhoods and improve overall health.

The target location for this analysis is Rochester, NY and its bedroom communities, located in Monroe County.  Rochester, and the top of Monroe County, sit along the shore of Lake Ontario, halfway between Buffalo and Syracuse.

Similar to the class labs, food venues within a mile radius will be retrieved, sorted and classified.  Using K-means clustering, the schools and their venues will be grouped to show similar values and mapped.  Each cluster will be examined for information on the types of food available to hungry students searching for a quick bite to eat.

A possible expansion to this data would be to include poverty/income data for the areas in which the school is located to allow further analysis.  This data is can be found on the US Census web site, but as of yet obtaining it electronically is elusive.  It will be included if possible.

There are a few assumptions being made in this analysis:
1. Students at the schools have the ability to leave campus on breaks or at lunch
2. A one mile distance allows walking to the venues, since we assume not all students have transportation.  
3. Some assumptions about menu will be made based on the type of venue.  For instance - "fast food" assumes a chain-type restaurant like McDonalds or Wendy's, while "cafe" assumes healthier fare, like sandwiches and soups.
